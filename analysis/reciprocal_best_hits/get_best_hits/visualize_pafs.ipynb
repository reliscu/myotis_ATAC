#### Plot results of mapping Myotis-Californicus to Myotis-Auriculus:

```r
working_paf <- read_paf(list.files("resources", pattern = ".paf$", full.names = TRUE)[1])
head(working_paf)
```

```r
dotplot(working_paf, xlab = "Auriculus (query)", ylab = "Californicus (query)")
```

```r
ggplot(working_paf, aes(alen)) +
    geom_histogram(colour = "black", fill = "steelblue", bins = 20) +
    theme_bw(base_size = 16) +
    ggtitle("Distribution of alignment lengths") +
    scale_x_continuous(trans="log10", labels=comma) +
    xlab("Alignment lengths") + ylab("")
```
