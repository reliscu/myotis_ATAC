```r
setwd("/Users/rebecca/sudmant/analyses/myotis/analysis/reciprocal_best_hits/01_get_best_hits")

library(pafr)
library(dplyr, quietly = TRUE)
library(data.table)
```

```r
hits <- read_paf("resources/pafs/mMyoAui1.cleaned.hapheader_mMyoCai1.cleaned.hapheader.exact.paf", include_tags = FALSE)
```

```r
head(hits)
```

```r
hits <- hits[order(hits$qname	, hits$qstart),]
```

```r

chromsize <- fread("/Users/rebecca/sudmant/analyses/myotis/data/mMyoCai_chromsizes", data.table = FALSE)
colnames(chromsize) <- c("Chr", "Chr_Length")
```

```r
hit_length <- hits %>%
  dplyr::mutate(
    Chr = sapply(strsplit(qname, ".", fixed = TRUE), "[", 3),
    Length = qend - qstart
  ) %>%
  dplyr::group_by(Chr) %>%
  dplyr::summarise(Recip_Length = sum(Length))

hit_length <- merge(chromsize, hit_length, by = 1, all = TRUE)
hit_length$Recip_Length[is.na(hit_length$Recip_Length)] <- 0
hit_length$Frac = with(hit_length, Recip_Length / Chr_Length)
hit_length$Total_Frac = with(hit_length, sum(Recip_Length)/sum(Chr_Length))
```
