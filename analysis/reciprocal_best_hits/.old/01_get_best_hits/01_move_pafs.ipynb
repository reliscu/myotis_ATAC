```r
setwd("/Users/rebecca/sudmant/analyses/myotis/analysis/reciprocal_best_hits")

file_list <- c()

data_dir <- "/Users/rebecca/sudmant/data/myotis/genomes/wfmash/-1-to-1"

myo_meta <- read.csv("../../data/myotis_meta.csv")
head(myo_meta)
```

```r
for (i in seq_along(1:length(myo_meta$Field_Name))) {
    for (j in seq_along(1:length(myo_meta$Field_Name))) {
        paf <- intersect(
            list.files(path = data_dir, pattern = paste0("^", myo_meta$Abbr[i])),
            list.files(path = data_dir, pattern = paste0("_", myo_meta$Abbr[j]))
            )
        file.copy(from = file.path(data_dir, paf), to = "resources/01_get_best_hits")
    }
}

list.files(path=file.path(getwd(), "resources/01_get_best_hits"))
```
