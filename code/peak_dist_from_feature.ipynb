```r
setwd("/Users/rebecca/sudmant/analyses/myotis/data/ATAC-seq")

library(data.table)
```

```r
## Median peak dist from genes

species <- list.files(path = "genes", full.names = TRUE)
gene_dist <- lapply(species, function(spec) {
  peaks <- fread(paste0(spec, "/Consensus_Peaks_Annotations_Accross_All_Samples.csv"), data.table = FALSE)
  return(median(peaks$`Distance to TSS`, na.rm = TRUE))
})
hist(unlist(gene_dist), breaks = 3)
```

```r
median(unlist(gene_dist))
```
